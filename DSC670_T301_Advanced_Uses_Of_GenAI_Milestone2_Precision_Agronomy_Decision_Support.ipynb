Step 2: Loading the Model with 4-bit Quantization
Approach to finally pull those 16GB files into your 16GB T4 GPU. By using load_in_4bit, we compress the model so it only takes up ~5.5GB of VRAM, leaving me room to actually run  Ag-Intel prompts.

```python
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer, BitsAndBytesConfig

model_id = "meta-llama/Llama-3.1-8B-Instruct"

# This config is the "magic" that makes a 16GB model fit on a T4 GPU
bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_use_double_quant=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.bfloat16
)

print("Initializing model on GPU...")

model = AutoModelForCausalLM.from_pretrained(
    model_id,
    quantization_config=bnb_config,
    device_map="auto"  # This automatically maps the model to your T4 GPU
)

# We already have the tokenizer from your previous step, but let's ensure it's ready
tokenizer = AutoTokenizer.from_pretrained(model_id)
tokenizer.pad_token = tokenizer.eos_token

print("\n--- Success! Ag-Intel Baseline Model is Ready ---")
```

Step 3: Running Milestone 2 Experiments

```python
def run_ag_test(query):
    messages = [
        {"role": "system", "content": "You are a professional agronomist."},
        {"role": "user", "content": query}
    ]
    prompt = tokenizer.apply_chat_template(messages, tokenize=False, add_generation_prompt=True)
    inputs = tokenizer(prompt, return_tensors="pt").to("cuda") # Moving input to GPU

    with torch.no_grad():
        outputs = model.generate(**inputs, max_new_tokens=256, temperature=0.7)

    return tokenizer.decode(outputs[0][inputs['input_ids'].shape[1]:], skip_special_tokens=True)

# Test 1: Groundnut Disease
print("Groundnut Test Results:")
print(run_ag_test("My groundnut crop shows yellowish-green patches and brown necrotic areas. Cause and treatment?"))
```

Why this is the "Fix":
CUDA Error: Using device_map="auto" inside from_pretrained forces the library to re-check for the GPU and should bypass that AssertionError.

Memory Management: Loading "raw" Llama 3.1 (16GB) on a T4 (15GB usable) will crash eventually. The BitsAndBytesConfig ensures you have enough overhead to finish the assignment without the session crashing mid-download.
